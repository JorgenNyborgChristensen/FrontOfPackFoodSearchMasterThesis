## Calculate WHO score for recipes
* A recipe or meal plan with a WHO score of 7 is interpreted as being very healthy whereas a score of 0 is seen as very unhealthy 
* FSA score: range from 4 (very healthy recipe) to 12 (very unhealthy recipe)

```python
#recipes = pd.read_csv("./Condition_1_nutriscore/recipes.csv")
# Correct FSA scores
recipes = pd.read_csv("recipes_correct_fsa.csv")
```

```python
def WHO( d_calories, d_protein, d_carbs, d_sugar, d_fat, d_saturated_fat, d_fiber, d_sodium):
    d_calories = d_calories
    d_protein = d_protein
    d_carbs = d_carbs
    d_sugar = d_sugar
    d_fat = d_fat
    d_saturated_fat = d_saturated_fat
    d_fiber = d_fiber
    d_sodium = d_sodium
    
    #carbs - sugar = total cals should be correct

    total = d_calories;
    prot = (d_protein*4.0)/total*100.0; #10-15 
    carbs = (d_carbs*4.0)/total*100.0; # 55-75
    sugar = (d_sugar*4.0)/total*100.0; # <10
    fat = (d_fat*9.0)/total*100.0; # 15-30
    sat_fat = (d_saturated_fat*9.0)/total*100.0; # <10
    fiber = (d_fiber)/(total/238.84); # denisty # >3.0
    sodium = (d_sodium)/(total/238.84);

    cnt = 0;
    if (prot >= 10.0 and prot <= 15.0): 
        cnt+= 1
    
    if (carbs >= 55.0 and carbs <= 75.0):
        cnt+= 1

    if (sugar < 10.0):
        cnt+= 1

    if (fat >= 15.0 and fat <= 30.0):
        cnt+= 1
        
    if (sat_fat < 10.0):
        cnt+= 1
    if (fiber > 3.0):
        cnt+= 1
    if (sodium < 0.2):
        cnt+= 1
        
    return cnt;
```

```python
recipes['sodium_per_100g'] = recipes['sodium_per_100g'].divide(1000) # Convert sodium to mg
#recipes['recipe_calories'] = recipes['recipe_calories'].str.strip('**') # Convert sodium to mg
#recipes['recipe_calories'] = recipes['recipe_calories'].astype(float)

recipes['who_score'] = recipes.apply(lambda row: WHO( row.kcal_per_100g, row.protein_per_100g, row.carbohydrates_per_100g, row.sugar_per_100g, row.fat_per_100g, row.saturated_fat_per_100g, row.fiber_per_100g, row.sodium_per_100g), axis = 1)
```
